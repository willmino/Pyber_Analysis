# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

import numpy as np

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type

rides_per_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

riders_df = rides_per_city_type.to_frame()

print(rides_per_city_type)
```

```python
# 2. Get the total drivers for each city type

#note that we use the city_data_df dataframe as the input dataframe for driver_count calculations
#and not the pyber_data_df dataframe. If the driver count sum statistic is taken from the pyber_data_df,
# you will instead get the driver_count for all the individual rides, and not the unique number of drivers

drivers_per_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]

drivers_df = drivers_per_city_type.to_frame()

drivers_df

print(drivers_per_city_type)
```

```python
#  3. Get the total amount of fares for each city type
fares_per_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]

fares_df = fares_per_city_type.to_frame()

fares_df

print(fares_per_city_type)
```

```python
#  4. Get the average fare per ride for each city type. 

#For each type of city, calculate the average fare per ride. 
#Which is (total fare per city)/(total rides per city)

average_fare_per_ride = fares_per_city_type/rides_per_city_type

fare_per_ride_df = average_fare_per_ride.to_frame()

fare_per_ride_df

print(average_fare_per_ride)
```

```python
# 5. Get the average fare per driver for each city type. 

average_fare_per_driver = fares_per_city_type/drivers_per_city_type

fare_per_driver_df = average_fare_per_driver.to_frame()

fare_per_driver_df

print(average_fare_per_driver)
```

```python
#  6. Create a PyBer summary DataFrame. 

#to create the dataframe, merge one at a time each column you would like to add

dfs = [drivers_df, fares_df, fare_per_ride_df, fare_per_driver_df]

pyber_summary_df = riders_df

for i in dfs:
    pyber_summary_df = pd.merge(pyber_summary_df,i, how = "left", on = "type")

pyber_summary_df = pyber_summary_df.rename(columns={"ride_id":"Total Rides", "driver_count":"Total Drivers", "fare":"Total Fares",
                                "0_x":"Average Fare per Ride","0_y":"Average Fare per Driver"})


pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None

```

```python
#  8. Format the columns.

columnlist = ["Total Fares","Average Fare per Ride","Average Fare per Driver"]

for i in columnlist:
    pyber_summary_df[i] = pyber_summary_df[i].map('${:,.2f}'.format)


columnlist2 = ["Total Rides","Total Drivers"]

for i in columnlist2:
    pyber_summary_df[i] = pyber_summary_df[i].map('{:,}'.format)


pyber_summary_df

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_summary_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_by_date = pyber_data_df.groupby(["type","date"]).sum()["fare"]

fare_by_date
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fare_by_date = fare_by_date.reset_index()

fare_by_date
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 


fare_by_date_pivot = fare_by_date.pivot(index = "date", columns = "type", values = "fare")

fare_by_date_pivot.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.

fare_by_date_pivot_df = fare_by_date_pivot.loc['2019-01-01':'2019-04-28']

fare_by_date_pivot_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fare_by_date_pivot_df.index = pd.to_datetime(fare_by_date_pivot_df.index)

```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fare_by_date_pivot_df.info()

fare_by_date_pivot_df
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

fare_by_week_pivot_df = fare_by_date_pivot_df.resample("W").sum()

fare_by_week_pivot_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


#pythonic way to get the number of months required for labels of

timestamps = fare_by_week_pivot_df.index
weeks = timestamps.astype(str).tolist()
months = []

for i in weeks:
    if i[5:7] not in months:
        months.append(i[5:7])
#the input for axis labels using the ax.set_xticks() function must be a tuple
xlabels = ("Jan\n2019", "Feb", "Mar", "Apr")

fig,ax = plt.subplots(figsize = (16,5.5))
                      
ax.plot(fare_by_week_pivot_df)

ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare($USD)")

lists = ax.get_xticks().tolist()
intlist = []
for i in lists:
    i = int(i)
    intlist.append(i)


start = min(intlist)

end = max(intlist)
increments = int((end-start)/len(months))
xaxis = np.arange(start,end,increments)
xvalues = [value for value in xaxis]
xaxis[0] = xaxis[0]+5

ax.set_xlim(737176,737177)



ax.set_xticks(xaxis)
ax.set(xticklabels = xlabels)


ax.legend(labels = ["Rural","Suburban", "Urban"], loc = "center", title = "type")

plt.tight_layout()


plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
